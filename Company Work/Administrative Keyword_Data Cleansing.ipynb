```python
# import modules
import pandas as pd
from datetime import datetime
```

```python
# combine administrative data sheet
for i in range(3):
    if i == 0:
        svc_total = pd.read_excel('C:\\Users\\paish\\OneDrive\\개인폴더\\Tensorflow\\생활맞춤형주제어01.xlsx', sheet_name = i)
    else:
        svc_sheet = pd.read_excel('C:\\Users\\paish\\OneDrive\\개인폴더\\Tensorflow\\생활맞춤형주제어01.xlsx', sheet_name = i)
        svc_total = pd.concat([svc_total, svc_sheet], ignore_index=True)

svc_total
```

```python
# 1. raw data cleansing ------------------------------------
svc_name = []
svc_content = []
ttl = []

# progress checking
cnt = 0

# data classification considering data pattern
for i in range(len(svc_total)):
    cnt += 1
    # notifying progress (per 100)
    if cnt % 100 == 0:
        print((cnt / len(svc_total)) * 100, " % 완료 되었어요!!")
        print('===================')
    
    if str(svc_total['<DOCID>'][i]).startswith('서비스명'):
        svc_name.append(svc_total['<DOCID>'][i])
    
    elif str(svc_total['<DOCID>'][i]).startswith('서비스내용'):
        # 'content' field needs seperate progress.
        content = svc_total['<DOCID>'][i]
        
        for j in range(i + 1, len(svc_total)):
            if not(svc_total['<DOCID>'][j].startswith('주제어')):
                content +=  str(svc_total['<DOCID>'][j])
                print(content)
            else:
                break
        print(content)
        svc_content.append(content)
    
    elif str(svc_total['<DOCID>'][i]).startswith('주제어'):
        ttl.append(svc_total['<DOCID>'][i])
    
    else:
        continue
#     print(i,' ',str(svc_total['<DOCID>'][i * 6 + 2]))
#     svc_name.append(str(svc_total['<DOCID>'][i * 6 + 2]))
#     svc_content.append(str(svc_total['<DOCID>'][i * 6 + 3]))
#     ttl.append(str(svc_total['<DOCID>'][i * 6 + 4]))
    
# remove unusable keyword
for i in range(len(svc_name)):
    svc_name[i] = svc_name[i][7:]
    svc_content[i] = svc_content[i][8:]
    ttl[i] = ttl[i][6:]
    
# checking data 
svc_name[len(svc_name) - 1]
```

```python
# 2. Converting Array to DataFrame
jeongbu24 = pd.DataFrame(list(zip(svc_name, svc_content, ttl)), columns = ['서비스명', '서비스내용', '주제어'])
jeongbu24
```

```python
# extract main field
jeongbu_main = jeongbu24['주제어']
print(len(jeongbu_main))
```

```python
ttl_dict = {}

# divide "Keyword - Score" into Dictionary
for i in range(len(jeongbu_main)):
#     print(jeongbu_main[i].split('^'))
    for j in range(len(jeongbu_main[i].split('^'))):
        ttl_dict[jeongbu_main[i].split('^')[j].split(':')[0]] = jeongbu_main[i].split('^')[j].split(':')[1]

# count dictionary
print(len(ttl_dict))
```

```python
# plus
import pandas as pd
data = pd.read_csv('C:\\Users\\paish\\Desktop\\통합 문서2.csv')
data
```

```python
from konlpy.tag import Okt

okt = Okt()

result = []

for i in range(len(data['keyword'])):
    if len(list(okt.phrases(data['keyword'][i]))) == 1:
        result.append('단일어')
    elif len(list(okt.phrases(data['keyword'][i]))) > 1:
        result.append('복합어')
    else:
        result.append('X')

data['result'] = result
```

```python
# 3. Export to Excel
exportFileName = 'C:\\Users\\paish\\Desktop\\ttlSave_' + datetime.today().strftime("%Y%m%d%H%M%S") + '.csv'

data.to_csv(exportFileName, encoding="utf-8-sig")
```
