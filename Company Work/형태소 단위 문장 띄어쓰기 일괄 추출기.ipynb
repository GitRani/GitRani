```python
from konlpy.tag import Okt
okt = Okt()

def spacing_okt(wrongSentence):
    tagged = okt.pos(wrongSentence)
    corrected = ""
    for i in tagged:
        if i[1] in ('Josa','Punctuation'):
            corrected += i[0]
        else:
            corrected += " "+i[0]
    if corrected[0] == " ":
        corrected = corrected[1:]
    return corrected

print(spacing_okt("아버지가방에들어가신다."))
```

```python
from eunjeon import Mecab
mecab = Mecab()

def spacing_mecab(wrongSentence):
    tagged = mecab.pos(wrongSentence)
    corrected = ""
    for i in tagged:
        # 'Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation',,'XSN','XPN', 'JKG', 'MM', 'JKB'
        # 한글자는 아예 붙여버리려고 함. 코드 체크 필요
        if len(i[0]) == 1:
            corrected += i[0]       
        elif i[1] in ('Josa', 'Punctuation'):
            corrected += i[0]
        elif i[1].startswith('X') or i[1].startswith('V') or i[1].startswith('E') or i[1].startswith('J'):
            corrected += i[0]
        elif i[1] in ('NNG'):
            corrected += " "+i[0]
        else:
            corrected += " "+i[0]
    if corrected[0] == " ":
        corrected = corrected[1:]
    return corrected

print(spacing_mecab("아버지가방에들어가신다."))
```

```python
import pandas as pd
from itertools import *
from datetime import datetime

importFileName = 'C:\\Users\\paish\\Desktop\\일괄 복합어 처리 - 1차정제' + '.csv'
exportFileName = 'C:\\Users\paish\Desktop\save_' + datetime.today().strftime("%Y%m%d%H%M%S") + '.csv'

blank_set = ['', ' ']
big_set = []

bokhap = pd.read_csv(importFileName)

for i in range(len(bokhap)):
    pos_set = []
    case_set = []
    text = ""
    print('===============',i,'==================')
    # 복합어의 조건
    if len(spacing_mecab(bokhap['검색어'][i]).split(' ')) > 1 :
        # 형태소를 모은다.
        pos_set = spacing_mecab(bokhap['검색어'][i]).split(' ')
        # 띄어쓰기 단위를 모은다.
        blank_list = list(product(blank_set, repeat = len(pos_set) - 1))
    
        print('pos_set', pos_set)
        for k in range(len(blank_list)):
            for j in range(len(pos_set)):
                text += pos_set[j]
                if j != len(pos_set):
                    text += blank_list[k][j - 1]
            case_set.append(text)
            text = ""
    else:
        print("탈출")
        
        
    print(case_set)
    big_set.append(case_set)
     
bokhap['분리결과'] = big_set
bokhap.to_csv(exportFileName, encoding="utf-8-sig") 
```
