```python
# 사이킷런의 데이터 셋을 불러온다.
import sklearn.datasets

# 많이 사용하는 패키지 import
# numpy : 배열을 관리하기 위한 패키지, pandas : 데이터프레임(tabular data set) 사용을 위한 패키지
# visual library : seaborn, matplotlib 으로 둘 다 많이 사용됨.
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

# visualize 결과물을 노트북에 보여주겠다.
%matplotlib inline
```

```python
data = sklearn.datasets.load_boston()
```

```python
print(data.keys())
```

```python
print(data.target[:5])
```

```python
print(data.DESCR)
```

```python
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df.head(5)
```

```python
df['house_price']= data.target
df.head(5)
```

```python
# 깔끔함. 하지만 값의 편차는 존재할듯.
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# 일부 열만 가져와서 그래프로 확인
# CRIM은 400 전후로 증가세 -> housing pricing 감소 (범죄율과 집값과의 관계가 있을 듯.)

df[['CRIM','ZN','INDUS','house_price']].plot(figsize = (20, 5))

plt.show()
```

```python
# AGE와도 연관성이 있을 것 같음.
df[['CHAS','NOX','RM','house_price','AGE']].plot(figsize = (20, 5))
plt.show()
```

```python
# AGE와도 연관성이 있을 것 같음.??
df[['DIS','RAD','PTRATIO','house_price']].plot(figsize = (20, 5))

plt.show()
```

```python
sns.pairplot(df[['CRIM','AGE','house_price']])
```

```python
#pairplot은 연속적인 값에는 사용할 수 없음..
section = []

for i in range(len(df)):
    if df['house_price'][i] < 15:
        section.append('low price')
    elif df['house_price'][i] >= 15 and df['house_price'][i] < 30:
        section.append('middle price')
    else:
        section.append('high price')
        
print(section[:10])

test_df = df
test_df['section'] = section
sns.pairplot(test_df[['CRIM','NOX','RM','section']], hue = 'section')
```
